```python
import cv2
import os


path_par = "/content/images/pares"
path_impar = "/content/images/impares"
path_salida = '/content/images/salidas'
archivos_par = os.listdir(path_par)
archivos_impar = os.listdir(path_impar)
print(archivos_par)
print(archivos_impar)

for archivo in archivos_par:
  image_par = cv2.imread(path_par + "/" + archivo, cv2.IMREAD_UNCHANGED)
  #image_impar = cv2.imread(path_impar
  imageOut_par = image_par[70:625,0:1105]
  cv2.imwrite(path_salida + "/" + archivo,imageOut_par)

for archivo in archivos_impar:
  image_impar = cv2.imread(path_impar  + "/" + archivo, cv2.IMREAD_UNCHANGED)
  imageOut_impar = image_impar[200:1425,60:950]
  cv2.imwrite(path_salida + "/" + archivo,imageOut_impar)
```

```python
import zipfile

archivos_salida = os.listdir(path_salida)

try:
    import zlib
    compression = zipfile.ZIP_DEFLATED
except:
    compression = zipfile.ZIP_STORED
zf = zipfile.ZipFile("/content/images/salidas/archivo_imagenes_compilados.zip", mode="w")
try:
    for file in archivos_salida:
      zf.write(path_salida + "/" +file, compress_type=compression)
    #zf.write("/content/sample_data/mnist_train_small.csv", compress_type=compression)
    #zf.write("/home/decodigo/Documentos/python/archivos/archivo3.txt", compress_type=compression)
    #zf.write("/home/decodigo/Documentos/python/archivos/archivo4.txt", compress_type=compression)
finally:
    zf.close()
```

```python
import os
import img2pdf

contador=10
imagenes_png = []
imagenes_png_dir = "/content/salida/"
for archivo in os.listdir(imagenes_png_dir):
  if archivo.endswith(".png"):
    #print(archivo)
    path = os.path.join(imagenes_png_dir, archivo)
    imagenes_png.append(path) 
print(imagenes_png)
with open("/content/salida/pdf/"+ str(contador)+".pdf", "wb") as documento:
  documento.write(img2pdf.convert(imagenes_png))
```
